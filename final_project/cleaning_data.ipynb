```python
import pandas as pd 
import geocoder
import time
import reverse_geocoder as rg
```

```python
df = pd.read_csv("2018_08_02_jessica_testset2.csv")
```

```python
df.head()
```

```python
len(df.lat)
```

```python
lat_lng_string =[]
state_confirmed =[]
```

```python
for a,b in zip(df.lat, df.lng): 
    lat_lng_string.append((a,b)) 
```

```python
for cord in lat_lng_string:
    results = rg.search(cord,mode=1)
    state_confirmed.append(((results)[0])['admin1'])
    print(((results)[0])['admin1'])
```

```python
len(state_confirmed)
```

```python
state = pd.Series(state_confirmed)
```

```python
df['confirmed_state'] = state.values
```

```python
df['confirmed_state'].unique()
```

```python
cleaned_df = df.loc[df['confirmed_state'] == 'Arizona']
```
